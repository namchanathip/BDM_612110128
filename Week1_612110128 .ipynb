```python
userInput = input("Serial number: ")
B = userInput.split(",")
even = ([i for item in B if int(item)%2 == 0 ])
odd = ([i for item in B if int(item)%2 != 0 ])
print("Number of even numbers : "+str(len(even)))
print("Number of odd numbers : "+str(len(odd)))
```
